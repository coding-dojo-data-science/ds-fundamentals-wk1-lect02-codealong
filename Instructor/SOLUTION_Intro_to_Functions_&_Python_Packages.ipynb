# Part 1: Dinner Party Entree Counts

## Our Task
>- You decided to throw a dinner party with 50 guests.
  -  The guests have a choice of one of the dinner options listed above.

- We will use np.random.choice() to simulate 50 random options that the guests attending the dinner party selected.
  - We will then create a for loop that tallies the number of entrees that the guests ordered during the dinner party.
  - We will then convert that loop into a function.
  <!-- - Finally, we will create a function using the for loop that was created to find the entree tallies for a dinner party with 150 guests.  -->

````python
# Using np.random.choice() to generate 50 random dinner option choices
selected_entrees = np.random.choice(dinner_options,50)

# Print the type and length of the result
print(type(selected_entrees))
print(len(selected_entrees))

# preview the first 5 values
selected_entrees[:5]
````

### Calculating Total Counts of Each Entree

- Create a for loop that uses a dictionary to tally the number occurances for each entree.

````python
# Create a for loop that tallies the counts of each entree

# Create an empty dictionary for entree counts
entree_counts = {}


# Loop through selected_entrees
for item in selected_entrees:


  # if the entree already exists in the dictionary:
  if item in entree_counts:
      
    # increase its count by 1
    entree_counts[item]+=1
    
  # if it doesn't already exist in the dictionary:
  else:
    # add the entree to the counts dictionary with value =1
    entree_counts[item] = 1

# display the final dictionary
entree_counts
````

- ***We want to be able to get this information for a new list of entrees, but how can do that without repeating the entire block of code with our loop?***

### Defining a `get_entree_counts` Functions

- Convert your loop from above into a function that you can reuse.

````python
# define your function here

def get_entree_counts(selected_items):
  
  # Your code goes here
  entree_counts = {}

  for item in selected_items:

    if item in entree_counts:
      
      # Two-Line Way
      # curr_num = entree_counts[item]
      # entree_counts[item] = curr_num +1

      # One-Line way
      entree_counts[item]+=1
    
    else: 
      entree_counts[item] = 1
    
  return entree_counts
````

````python
# test your function and save the output as "results"

results = get_entree_counts(selected_entrees)
results
````

# Part 2: Its Dinner Party Season! 

- Now, imagine we are planning several monthly dinner parties over the next 6 months. We have sent invites with our entree choices and all of our guests have RSVP'd with their selected entrees.


- We are planning our food orders for the next few months and we want to know the totals for each month's party.

>Run the cell below to create the dictionary with the guests's entree choices for every party.

````python
## Run this cell without changes

# Setting up the party months and number of guests
planned_parties = ['September','October','November','December',
                   "January", "February"]
n_guests = [28, 42, 53, 23, 32, 75]


## Creating a dictionary with a list of entrees for each month's party
monthly_entrees = {}

for i, month in enumerate(planned_parties):
  

  ## randomly select the right number of options
  entree_choices = np.random.choice(dinner_options, n_guests[i])

  ## save the entrees in the dict using the month as the key
  monthly_entrees[month] = entree_choices


# monthly_entrees
````

- Let's inspect this monthly_entrees variable:

````python
# what is monthly_entrees? what type of var?
type(monthly_entrees)
````

````python
# what do the values for September look like? what type of var?
monthly_entrees['September']
````

- We have a dictionary with the month name as the keys and the selected entrees as the values.


### Using Our Function in a Loop

Now that we have a dictionary of with the party month as the keys and the array of guest entrees as the values, we want to:

- Loop through our `monthly_entrees` dictionary


  - For each month we want to:

      - Use our helper function to get a dictionary of the # of entree counts for that month
      - We then want to print the month's name.
      - print the # of total sales for the entire month (length of the list of selected entrees)

      - and print entree counts from  selected for that month.

````python
## MVP solution

# for each month and its entree choices:
for month_name, entrees in monthly_entrees.items():

  # use get_entree_counts function to get the results for the current month
  entree_counts = get_entree_counts(entrees)

  # print month name
  print(f"For the month of {month_name}:")

  # print # of entrees total
  print(f"- there are {len(entree_counts)} entrees total.")

  # print the counts of each entree for the month
  print(entree_counts)
````

````python
## AAB soluton
# for each month and its entree choices:
for month_name, entrees in monthly_entrees.items():

  # use get_entree_counts function to get the results for the current month
  entree_counts = get_entree_counts(entrees)

  # print month name
  print(f"For the month of {month_name}:")

  # print # of entrees total
  print(f"There are {len(entrees)} entrees total.")

  # print the counts of each entree for the month
  for entree, count in entree_counts.items():
    print(f"- {entree}: {count}")

  print()
````
